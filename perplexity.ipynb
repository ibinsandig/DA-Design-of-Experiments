## 1. Einleitung

### 1.1 Relevanz des Themas

Das Design of Experiments (DoE) ist eine systematische Methode zur Planung, Durchführung und Auswertung kontrollierter Experimente. Sie ermöglicht es, die Auswirkungen mehrerer Einflussfaktoren gleichzeitig zu untersuchen und dabei den experimentellen Aufwand zu minimieren. Im Gegensatz zu einfaktoriellen Experimenten, bei denen jeweils nur eine Variable variiert wird, erlaubt das faktorielle Experimentdesign die effiziente Untersuchung von Wechselwirkungen zwischen Faktoren.

Die Analyse von Haupteffekten und Wechselwirkungen mittels ANOVA (Analysis of Variance) ist dabei zentral, um statistisch gesicherte Aussagen über die Signifikanz von Einflussfaktoren zu treffen.

### 1.2 Anwendungsbereiche

DoE wird in zahlreichen Industriebereichen angewendet:

- **Materialwissenschaften & Bauwesen:** Optimierung von Betonmischungen zur Maximierung der Druckfestigkeit
- **Pharmazie:** Entwicklung von Arzneimitteln und Optimierung von Wirkstoffen
- **Chemische Verfahrenstechnik:** Optimierung von Reaktionsbedingungen
- **Qualitätskontrolle:** Reduktion von Fertigungstoleranzen und Verbesserung der Produktqualität
- **Maschinenbau:** Optimierung von Produktionsprozessen
- **Elektrotechnik:** Verbesserung von Chip-Design und Halbleiterherstellung

## 2. Theoretische Grundlagen

### 2.1 Grundkonzepte des Design of Experiments

Das Design of Experiments basiert auf folgenden Kernprinzipien:

**Faktor :** Eine variable Eingangsgröße, deren Einfluss untersucht werden soll (z. B. Temperatur, Konzentration, Zeit).

**Level :** Die konkrete Ausprägung, die ein Faktor im Experiment annimmt (z. B. "niedrig" = -1, "hoch" = +1).

**Antwort :** Die gemessene Ausgangsgröße oder das Zielkriterium des Experiments (z. B. Druckfestigkeit).

**Wechselwirkung :** Die Situation, in der der Effekt eines Faktors von der Ausprägung eines anderen Faktors abhängt.

### 2.2 Faktorielles Experimentdesign

#### Vollfaktorielles Design (Full Factorial Design)
Bei einem vollfaktoriellen Design werden alle möglichen Kombinationen der Faktor-Level untersucht. Für \(k\) Faktoren mit jeweils \(n\) Levels ergeben sich \(n^k\) Versuchskombinationen.

**Vorteile:**
- Vollständige Information über alle Haupteffekte und Wechselwirkungen
- Keine Konfundierung von Effekten

**Nachteile:**
- Sehr hohe Anzahl von Experimenten bei vielen Faktoren
- Kostenintensiv und zeitaufwendig

#### Teilfaktorielles Design (Fractional Factorial Design)
Ein Teilfaktorielles Design ist eine Untermenge eines vollfaktoriellen Designs, bei dem nur ein Bruchteil (z. B. 1/2, 1/4) aller Kombinationen durchgeführt wird.

**Vorteile:**
- Deutlich weniger Versuchskombinationen
- Wirtschaftlicher und schneller

**Nachteile:**
- Gewisse Wechselwirkungen können konfundiert sein
- Weniger Informationen verfügbar

### 2.3 Mathematisches Modell

Das lineare Regressionsmodell für ein faktorielles Design mit zwei Faktoren A und B lautet:

\[
y = \mu + \alpha_i + \beta_j + (\alpha\beta)_{ij} + \epsilon_{ijr}
\]

Dabei ist:
- \(y\): Die gemessene Antwortvariable (z. B. Druckfestigkeit)
- \(\mu\): Der Gesamtmittelwert
- \(\alpha_i\): Der Haupteffekt von Faktor A auf Level i
- \(\beta_j\): Der Haupteffekt von Faktor B auf Level j
- \((\alpha\beta)_{ij}\): Die Wechselwirkung zwischen A und B
- \(\epsilon_{ijr}\): Der Fehlerterm (Residuum) der r-ten Replikation

**Haupteffekt:** Der durchschnittliche Einfluss eines Faktors auf die Antwortvariable, unabhängig von anderen Faktoren.

**Wechselwirkungseffekt:** Der zusätzliche Effekt, der entsteht, wenn zwei oder mehr Faktoren kombiniert sind.

### 2.4 ANOVA (Analysis of Variance)

ANOVA ist ein statistisches Verfahren zur Zerlegung der Gesamtvarianz in verschiedene Komponenten:

\[
SS_{Total} = SS_A + SS_B + SS_{AB} + SS_E
\]

Dabei ist:
- \(SS_{Total}\): Gesamtsumme der Quadrate
- \(SS_A, SS_B\): Summe der Quadrate für Faktor A und B (Haupteffekte)
- \(SS_{AB}\): Summe der Quadrate für die Wechselwirkung
- \(SS_E\): Fehlerquadratsumme (Residuen)

**F-Statistik:**

\[
F = \frac{MS_{Effekt}}{MS_E} = \frac{SS_{Effekt} / df_{Effekt}}{SS_E / df_E}
\]

Die F-Statistik wird verwendet, um zu testen, ob ein Effekt statistisch signifikant ist. Ist \(F > F_{\alpha, df_1, df_2}\), wird der Effekt als signifikant betrachtet.

**P-Wert:** Gibt an, mit welcher Wahrscheinlichkeit der beobachtete Effekt rein zufällig auftritt. Typisch gilt: \(p < 0.05\) bedeutet statistisch signifikant.

### 3.2 Experimentdesign Definition

Für die Analyse wählen wir die 4 wichtigsten Faktoren basierend auf domänenspezifischem Wissen aus:

1. **Cement** - Hauptbindemittel, direkter Einfluss auf Festigkeit
2. **Water** - Beeinflusst Hydrationsprozess und Porenstruktur
3. **Superplastic** - Verbessert Verarbeitbarkeit und kann Festigkeit erhöhen
4. **Age** - Aushärtungszeit, direkter Einfluss auf Festigkeit

Diese vier Faktoren bilden die Basis für unser faktorielles Design. Wir werden die Daten zunächst kategorisieren, um diskrete Levels zu erstellen, und dann ein Analyse-Design durchführen.


```python
# Kategorisierung der Faktoren in "Low" (-1) und "High" (+1)
df_doe = df.copy()

# Median-basierte Kategorisierung
factors_to_categorize = ['cement', 'water', 'superplastic', 'age']

for factor in factors_to_categorize:
    median = df_doe[factor].median()
    df_doe[f'{factor}_level'] = df_doe[factor].apply(lambda x: -1 if x < median else 1)

print("Kategorisierte Faktoren (Low = -1, High = +1):")
print(df_doe[['cement', 'cement_level', 'water', 'water_level', 
              'superplastic', 'superplastic_level', 'age', 'age_level', 'strength']].head(10))

print("\n" + "=" * 80)
print("EXPERIMENTDESIGN: Faktor-Level")
print("=" * 80)
for factor in factors_to_categorize:
    low = df_doe[df_doe[f'{factor}_level'] == -1][factor].mean()
    high = df_doe[df_doe[f'{factor}_level'] == 1][factor].mean()
    print(f"{factor.upper()}: Low = {low:.2f}, High = {high:.2f}")
```

### 3.3 Haupteffekte-Analyse

Der Haupteffekt eines Faktors wird als Differenz zwischen dem Mittelwert der Response bei hohem Level und dem Mittelwert bei niedrigem Level berechnet:

\[
\text{Haupteffekt}_{\text{Faktor}} = \bar{y}_{\text{high}} - \bar{y}_{\text{low}}
\]

Ein positiver Effekt bedeutet, dass eine Erhöhung des Faktors die Response erhöht, ein negativer Effekt das Gegenteil.


```python
# Berechnung der Haupteffekte
main_effects = {}

for factor in factors_to_categorize:
    low_mean = df_doe[df_doe[f'{factor}_level'] == -1]['strength'].mean()
    high_mean = df_doe[df_doe[f'{factor}_level'] == 1]['strength'].mean()
    effect = high_mean - low_mean
    
    main_effects[factor] = {
        'low_mean': low_mean,
        'high_mean': high_mean,
        'effect': effect
    }

print("=" * 80)
print("HAUPTEFFEKTE")
print("=" * 80)
effects_df = pd.DataFrame(main_effects).T
print(effects_df.round(3))

print("\n" + "-" * 80)
print("Interpretation der Haupteffekte:")
print("-" * 80)
for factor, values in main_effects.items():
    effect = values['effect']
    direction = "POSITIV" if effect > 0 else "NEGATIV"
    print(f"{factor.upper()}: {effect:+.2f} MPa ({direction})")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.ravel()

for idx, factor in enumerate(factors_to_categorize):
    factor_level_col = f'{factor}_level'
    
    low_data = df_doe[df_doe[factor_level_col] == -1]['strength']
    high_data = df_doe[df_doe[factor_level_col] == 1]['strength']
    
    # Box Plot
    data_to_plot = [low_data, high_data]
    bp = axes[idx].boxplot(data_to_plot, labels=['Low (-1)', 'High (+1)'], patch_artist=True)
    
    # Färbung
    for patch in bp['boxes']:
        patch.set_facecolor('lightblue')
    
    # Mittelwerte als Punkte
    means = [low_data.mean(), high_data.mean()]
    axes[idx].plot([1, 2], means, 'ro-', linewidth=2, markersize=8, label='Mittelwert')
    
    axes[idx].set_title(f'Haupteffekt: {factor.upper()}', fontsize=12, fontweight='bold')
    axes[idx].set_ylabel('Druckfestigkeit (MPa)', fontsize=11)
    axes[idx].grid(True, alpha=0.3)
    axes[idx].legend()
    
    # Effekt-Wert anzeigen
    effect = main_effects[factor]['effect']
    axes[idx].text(0.5, 0.95, f'Effekt: {effect:+.2f} MPa', 
                   transform=axes[idx].transAxes, ha='center', va='top',
                   bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5),
                   fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

### 3.4 ANOVA-Analyse (Analysis of Variance)

Die ANOVA-Analyse testet die statistische Signifikanz der Haupteffekte. Die Hypothesen lauten:

**H₀** (Nullhypothese): Der Faktor hat keinen signifikanten Einfluss auf die Response.
**H₁** (Alternativhypothese): Der Faktor hat einen signifikanten Einfluss auf die Response.

Die ANOVA-Tabelle zeigt die Summen der Quadrate (SS), Freiheitsgrade (DF), mittleren Quadrate (MS), F-Statistik und p-Wert. Ein p-Wert < 0.05 deutet auf statistische Signifikanz hin (95% Konfidenzlevel).


```python
# ANOVA mit Regressionsmodell
formula = 'strength ~ C(cement_level) + C(water_level) + C(superplastic_level) + C(age_level)'
model = ols(formula, data=df_doe).fit()

# ANOVA-Tabelle
anova_table = anova_lm(model, typ=2)

print("\n" + "=" * 80)
print("ANOVA-TABELLE (Typ II)")
print("=" * 80)
print(anova_table.round(4))

print("\n" + "-" * 80)
print("MODELLZUSAMMENFASSUNG")
print("-" * 80)
print(f"R² (Bestimmtheitsmaß): {model.rsquared:.4f}")
print(f"Adjusted R²: {model.rsquared_adj:.4f}")
print(f"F-Statistik: {model.fvalue:.4f}")
print(f"P-Wert des Modells: {model.f_pvalue:.6f}")
```

```python
alpha = 0.05
print("\n" + "=" * 80)
print("SIGNIFIKANZANALYSE (α = 0.05)")
print("=" * 80)

significant_factors = []
for factor in factors_to_categorize:
    p_value = anova_table.loc[f'C({factor}_level)', 'PR(>F)']
    is_significant = p_value < alpha
    significant_factors.append((factor, p_value, is_significant))
    
    status = "✓ SIGNIFIKANT" if is_significant else "✗ nicht signifikant"
    print(f"{factor.upper():15s}: p-Wert = {p_value:.6f} {status}")

print("\n" + "-" * 80)
print("Signifikante Faktoren (p < 0.05):")
print("-" * 80)
sig_factors = [f[0] for f in significant_factors if f[2]]
if sig_factors:
    for factor in sig_factors:
        print(f"  • {factor.upper()}")
else:
    print("  Keine signifikanten Faktoren bei α = 0.05")
```

### 3.5 Wechselwirkungsanalyse (Interaction Effects)

Eine Wechselwirkung (Interaction) tritt auf, wenn der Effekt eines Faktors von der Ausprägung eines anderen Faktors abhängt. Graphisch wird eine Wechselwirkung durch nicht-parallele Linien in einem Interaktionsplot angezeigt.

Das Zwei-Faktor-Wechselwirkungsmodell wird wie folgt formuliert:

\[
y = \mu + \alpha_i + \beta_j + (\alpha\beta)_{ij} + \epsilon
\]

Die Wechselwirkungssummen der Quadrate werden separat berechnet und in der erweiterten ANOVA-Tabelle berücksichtigt.


```python
# ANOVA mit Wechselwirkungstermen
formula_interaction = 'strength ~ C(cement_level) * C(water_level) + C(superplastic_level) * C(age_level)'
model_interaction = ols(formula_interaction, data=df_doe).fit()

# ANOVA-Tabelle mit Wechselwirkungen
anova_interaction = anova_lm(model_interaction, typ=2)

print("\n" + "=" * 80)
print("ANOVA-TABELLE MIT WECHSELWIRKUNGEN (Typ II)")
print("=" * 80)
print(anova_interaction.round(4))

print("\n" + "-" * 80)
print("MODELLVERGLEICH")
print("-" * 80)
print(f"Modell ohne Wechselwirkungen - R²: {model.rsquared:.4f}")
print(f"Modell mit Wechselwirkungen - R²: {model_interaction.rsquared:.4f}")
print(f"Verbesserung: {(model_interaction.rsquared - model.rsquared):.4f}")
```

```python
# Interaktionsplot: Cement × Water
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Cement × Water
cement_groups = df_doe.groupby('cement_level')
water_levels = sorted(df_doe['water_level'].unique())

for cement_level in sorted(df_doe['cement_level'].unique()):
    cement_data = df_doe[df_doe['cement_level'] == cement_level]
    water_means = [cement_data[cement_data['water_level'] == wl]['strength'].mean() 
                   for wl in water_levels]
    label = f"Cement: High (+1)" if cement_level == 1 else f"Cement: Low (-1)"
    axes[0].plot(water_levels, water_means, 'o-', linewidth=2.5, markersize=8, label=label)

axes[0].set_xlabel('Water Level', fontsize=11, fontweight='bold')
axes[0].set_ylabel('Druckfestigkeit (MPa)', fontsize=11, fontweight='bold')
axes[0].set_title('Wechselwirkung: Cement × Water', fontsize=12, fontweight='bold')
axes[0].set_xticks([-1, 1])
axes[0].set_xticklabels(['Low', 'High'])
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Superplastic × Age
age_levels = sorted(df_doe['age_level'].unique())

for sp_level in sorted(df_doe['superplastic_level'].unique()):
    sp_data = df_doe[df_doe['superplastic_level'] == sp_level]
    age_means = [sp_data[sp_data['age_level'] == al]['strength'].mean() 
                 for al in age_levels]
    label = f"Superplastic: High (+1)" if sp_level == 1 else f"Superplastic: Low (-1)"
    axes[1].plot(age_levels, age_means, 's-', linewidth=2.5, markersize=8, label=label)

axes[1].set_xlabel('Age Level', fontsize=11, fontweight='bold')
axes[1].set_ylabel('Druckfestigkeit (MPa)', fontsize=11, fontweight='bold')
axes[1].set_title('Wechselwirkung: Superplastic × Age', fontsize=12, fontweight='bold')
axes[1].set_xticks([-1, 1])
axes[1].set_xticklabels(['Low', 'High'])
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
print("\n" + "=" * 80)
print("WECHSELWIRKUNGSEFFEKTE (Zwei-Faktor)")
print("=" * 80)

interactions = [
    ('cement', 'water'),
    ('cement', 'superplastic'),
    ('cement', 'age'),
    ('water', 'superplastic'),
    ('water', 'age'),
    ('superplastic', 'age')
]

interaction_effects = {}

for factor1, factor2 in interactions:
    level1_col = f'{factor1}_level'
    level2_col = f'{factor2}_level'
    
    # Vier Kombinationen
    hh = df_doe[(df_doe[level1_col] == 1) & (df_doe[level2_col] == 1)]['strength'].mean()  # High-High
    ll = df_doe[(df_doe[level1_col] == -1) & (df_doe[level2_col] == -1)]['strength'].mean()  # Low-Low
    hl = df_doe[(df_doe[level1_col] == 1) & (df_doe[level2_col] == -1)]['strength'].mean()  # High-Low
    lh = df_doe[(df_doe[level1_col] == -1) & (df_doe[level2_col] == 1)]['strength'].mean()  # Low-High
    
    # Wechselwirkungseffekt
    interaction_effect = ((hh + ll) - (hl + lh)) / 2
    interaction_effects[f"{factor1} × {factor2}"] = interaction_effect

# Sortieren nach Betrag des Effekts
sorted_interactions = sorted(interaction_effects.items(), key=lambda x: abs(x[1]), reverse=True)

print(f"\n{'Wechselwirkung':<25} {'Effekt (MPa)':>15}")
print("-" * 41)
for interaction, effect in sorted_interactions:
    print(f"{interaction:<25} {effect:>15.3f}")
```

### 3.6 Residuenanalyse

Die Residuenanalyse überprüft die Gültigkeit der ANOVA-Annahmen:
- Normalverteilung der Residuen
- Homogenität der Varianzen
- Unabhängigkeit der Residuen


```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Residuen vs. Fitted Values
fitted_values = model.fittedvalues
residuals = model.resid

axes[0, 0].scatter(fitted_values, residuals, alpha=0.6, s=50)
axes[0, 0].axhline(y=0, color='r', linestyle='--', linewidth=2)
axes[0, 0].set_xlabel('Fitted Values', fontsize=11, fontweight='bold')
axes[0, 0].set_ylabel('Residuals', fontsize=11, fontweight='bold')
axes[0, 0].set_title('Residuen vs. Fitted Values', fontsize=12, fontweight='bold')
axes[0, 0].grid(True, alpha=0.3)

# 2. Q-Q Plot (Normalverteilung)
stats.probplot(residuals, dist="norm", plot=axes[0, 1])
axes[0, 1].set_title('Q-Q Plot (Normalverteilung)', fontsize=12, fontweight='bold')
axes[0, 1].grid(True, alpha=0.3)

# 3. Histogramm der Residuen
axes[1, 0].hist(residuals, bins=15, edgecolor='black', alpha=0.7, color='skyblue')
axes[1, 0].set_xlabel('Residuals', fontsize=11, fontweight='bold')
axes[1, 0].set_ylabel('Häufigkeit', fontsize=11, fontweight='bold')
axes[1, 0].set_title('Verteilung der Residuen', fontsize=12, fontweight='bold')
axes[1, 0].grid(True, alpha=0.3, axis='y')

# 4. Scale-Location Plot
standardized_residuals = residuals / np.std(residuals)
axes[1, 1].scatter(fitted_values, np.sqrt(np.abs(standardized_residuals)), alpha=0.6, s=50)
axes[1, 1].set_xlabel('Fitted Values', fontsize=11, fontweight='bold')
axes[1, 1].set_ylabel('√|Standardisierte Residuen|', fontsize=11, fontweight='bold')
axes[1, 1].set_title('Scale-Location Plot', fontsize=12, fontweight='bold')
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Normalitätstest (Shapiro-Wilk)
statistic, p_value = stats.shapiro(residuals)
print("\n" + "=" * 80)
print("RESIDUENANALYSE: NORMALITÄTSTEST (Shapiro-Wilk)")
print("=" * 80)
print(f"Teststatistik: {statistic:.4f}")
print(f"P-Wert: {p_value:.6f}")
if p_value > 0.05:
    print("✓ Residuen sind normalverteilt (H₀ nicht verworfen, α = 0.05)")
else:
    print("✗ Residuen sind nicht normalverteilt (H₀ verworfen, α = 0.05)")
```

### 3.7 Zusammenfassung der Ergebnisse

Auf Basis der ANOVA-Analyse werden die signifikanten Faktoren identifiziert und quantifiziert. Die Haupteffekte zeigen die Größe des Einflusses, während die Wechselwirkungsanalyse aufdeckt, ob Faktoren synergistisch wirken.


```python
print("\n" + "=" * 80)
print("FAKTOR-EFFEKTE: RANGLISTE (nach Betrag)")
print("=" * 80)

effect_ranking = []
for factor in factors_to_categorize:
    effect = main_effects[factor]['effect']
    p_value = anova_table.loc[f'C({factor}_level)', 'PR(>F)']
    effect_ranking.append({
        'Faktor': factor.upper(),
        'Effekt (MPa)': effect,
        'Betrag |Effekt|': abs(effect),
        'P-Wert': p_value,
        'Signifikanz': '✓' if p_value < 0.05 else '✗'
    })

ranking_df = pd.DataFrame(effect_ranking).sort_values('Betrag |Effekt|', ascending=False)
print(ranking_df[['Faktor', 'Effekt (MPa)', 'P-Wert', 'Signifikanz']].to_string(index=False))
```

## 4. Diskussion und Fazit

### 4.1 Bewertung der Methode

Das faktorielle Experimentdesign hat sich als effektive Methode zur systematischen Untersuchung von Einflussfaktoren erwiesen. Die Kombination von kategorisierten Faktoren und ANOVA-Analyse ermöglichte es uns, die statistisch signifikanten Einflussfaktoren auf die Betondruckfestigkeit zu identifizieren.

**Stärken des Ansatzes:**
- Systematische und reproduzierbare Analyse
- Gleichzeitige Betrachtung mehrerer Faktoren
- Quantifizierung von Haupteffekten und Wechselwirkungen
- Klare statistische Tests zur Signifikanzbestimmung
- Effiziente Nutzung von Versuchsdaten

**Praktische Erkenntnisse:**
Aus der Analyse lässt sich ableiten, welche Faktoren bei der Betonmischungsoptimierung Priorität haben sollten. Faktoren mit großen positiven Haupteffekten und statistischer Signifikanz sollten in Optimierungsexperimenten gezielt variiert werden.

### 4.2 Grenzen der Analyse

1. **Kategorisierung der Faktoren:** Die Einteilung in nur zwei Level (Low/High) anhand des Medians vereinfacht die Realität. In der Praxis sind kontinuierliche oder mehrwertige Faktoren oft informativer.

2. **Limitierte Datenmenge:** Der Analysedatensatz umfasst nur 13 Beobachtungen. Für robustere Ergebnisse wären mehr Daten wünschenswert.

3. **Lineare Modellannahmen:** Das verwendete Modell geht von linearen Effekten aus. Nicht-lineare Effekte (z. B. polynomisch) werden nicht erfasst.

4. **Höherordnungsinteraktionen:** Diese Analyse betrachtet nur Zwei-Faktor-Wechselwirkungen. Drei- oder höherordnungsinteraktionen werden nicht analysiert.

5. **Verallgemeinerbarkeit:** Die Erkenntnisse beziehen sich auf diesen spezifischen Datensatz. Die Generalisierbarkeit auf andere Betonformulierungen ist begrenzt.

### 4.3 Ausblick und Empfehlungen

**Weiterführende Analysen:**
- Durchführung eines Response Surface Designs (RSD) für optimale Faktoreinstellungen
- Integration von Machine Learning-Modellen (z. B. Random Forest, Neural Networks) für nicht-lineare Muster
- Validierung der Ergebnisse mit neuen experimentellen Daten
- Analyse von Kosten-Nutzen-Verhältnissen bei optimierten Mischungen

**Praktische Anwendung:**
Die identifizierten Haupteffekte können als Grundlage für die Optimierung von Betonmischungen verwendet werden. Faktoren mit starken positiven Effekten sollten erhöht werden, während solche mit negativen Effekten reduziert werden sollten - unter Berücksichtigung von Kosten- und Verfügbarkeitsbeschränkungen.

**Schlussfolgerung:**
Das Design of Experiments mit ANOVA-Analyse hat sich als wertvolles Werkzeug zur Identifikation und Quantifizierung von Einflussfaktoren erwiesen. Die Kombination von systematischer Planung, statistischer Analyse und praktischer Interpretation ermöglicht datengetriebene Entscheidungen in komplexen technischen Systemen.
