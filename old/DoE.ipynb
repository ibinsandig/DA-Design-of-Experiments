Einordnung und Ziel von DoE

Design of Experiments (DoE) bezeichnet einen statistischen Ansatz zur geplanten, systematischen Durchführung von Experimenten, um den Einfluss mehrerer Eingangsgrößen (Faktoren) auf eine Zielgröße (Response) effizient zu untersuchen. Im Gegensatz zur Einfaktor‑Methode (One-Factor-at-a-Time, OFAT) werden bei DoE mehrere Faktoren gleichzeitig variiert, wodurch Haupteffekte und Wechselwirkungen quantifiziert und mit vergleichsweise wenigen Versuchen geschätzt werden können.


Mathematische Grundlage und Modelle

In der DoE‑Datenanalyse wird typischerweise ein lineares Regressionsmodell für die Zielgröße YY verwendet, etwa für ein zweistufiges 2k2k-Faktoriell‑Design:
Y=β0+∑i=1kβixi+∑i<jβijxixj+εY=β0+∑i=1kβixi+∑i<jβijxixj+ε.
​
Die Koeffizienten βiβi repräsentieren Haupteffekte der Faktoren, während βijβij die Wechselwirkungen beschreiben; der Fehlerterm εε wird als Zufallsvariable mit Erwartungswert 0 und konstanter Varianz σ2σ2 angenommen.

​

Zur Signifikanzprüfung der Effekte wird meist eine Varianzanalyse (ANOVA) verwendet, bei der die Gesamtvarianz in Anteile „durch das Modell erklärt“ und „Fehlervarianz“ zerlegt wird. Aus den mittleren Quadratsummen (Mean Squares) werden F‑Statistiken berechnet, um Hypothesen der Form „Effekt = 0“ gegen „Effekt ≠ 0“ bei vorgegebenem Signifikanzniveau (z. B. α=0,05α=0,05) zu testen.
​​


Zentrale Begriffe

    Faktor: Einflussgröße, die im Experiment gezielt variiert wird (z. B. Temperatur, Druck).

​

Faktorstufe: Konkreter Ausprägungswert eines Faktors (z. B. Temperatur = 60 °C / 80 °C).

​

Response/Zielgröße: Messbare Ergebnisgröße des Experiments (z. B. Produktqualität, Durchsatz).

​

Haupteffekt: Änderung der mittleren Response beim Wechsel der Stufe eines Faktors, gemittelt über alle Stufen der anderen Faktoren.

​

Wechselwirkung: Situation, in der der Effekt eines Faktors von der Stufe eines anderen Faktors abhängt.

​

Vollfaktorielles Design: Alle möglichen Kombinationen der Faktorstufen werden experimentell realisiert (z. B. 2k2k-Design mit zwei Stufen pro Faktor).

​

Teilfaktorielles Design (fractional factorial): Nur ein Bruchteil der Kombinationen wird untersucht, um Ressourcen zu sparen, auf Kosten eines kontrollierten Alias‑Musters

Vollfaktoriell vs. teilfaktoriell

In einem vollfaktoriellen 2k2k-Design können sämtliche Haupteffekte und alle Wechselwirkungen bis zur Ordnung kk unabhängig geschätzt werden, vorausgesetzt, es liegen genügend Replikationen zur Schätzung der Fehlervarianz vor. Der Versuchsaufwand wächst jedoch exponentiell mit der Anzahl der Faktoren, sodass bei vielen Faktoren schnell sehr große Stichprobenumfänge entstehen.

​

Teilfaktorielle Designs (z. B. 2k−p2k−p) reduzieren die Zahl der Versuche, indem gezielt nur ein „Ausschnitt“ des vollständigen Faktorraums untersucht wird und bestimmte Effekte aliasiert werden. Sie eignen sich besonders für Screening‑Experimente, in denen zunächst die wenigen wirklich wichtigen Faktoren identifiziert werden sollen, bevor in späteren Experimenten detailliertere Modelle (z. B. mit Quadrattermen oder Response‑Surface‑Designs) geschätzt werden.

ANOVA für DoE

Die ANOVA zerlegt die Gesamtquadratsumme SSTotalSSTotal in Beiträge für Faktoren, Wechselwirkungen und Fehler, beispielsweise SSTotal=SSA+SSB+SSAB+SSErrorSSTotal=SSA+SSB+SSAB+SSError in einem zweifaktoriellen Design. Zu jedem Quadratsummenanteil werden Freiheitsgrade, mittlere Quadratsummen und F‑Werte bestimmt, die mit kritischen F‑Verteilungswerten verglichen oder über p‑Werte interpretiert werden.
​

​

Grafische Werkzeuge wie Haupteffektdiagramme und Wechselwirkungsdiagramme unterstützen die Interpretation, indem sie Trends und Interaktionseffekte visuell darstellen. Ergänzend wird häufig eine Residualanalyse durchgeführt, um Modellannahmen (Normalverteilung, Homoskedastizität, Unabhängigkeit) zu prüfen und ggf. Transformationen oder alternative Modelle abzuleiten.
